```python
# to concat 2 data frames
df1 = pd.DataFrame({
    "name" : ["Akash","Ram","Shyam"],
    "marks": [3,100,12]
})
df2 = pd.DataFrame({
    "name" : ["Abhishek","Puja","Karan"],
    "marks": [311,1010,11]
})
print(pd.concat([df1,df2]))
```

```python
df1 = pd.DataFrame({
    "name" : ["Akash","Priyanka","Harry"],
    "marks" : [100,53,21]
})
df2 = pd.DataFrame({
    "name" : ["Akash","Priyanka"],
    "roll" : [20,21]
})
print(pd.merge(df1,df2,on="name")) # merge on name only; if name was there on one but not on other it will not show
```
